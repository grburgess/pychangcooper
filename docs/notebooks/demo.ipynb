# Generic Cooling and Acceleration

This demonstrates the a generic heating and cooling solution for 

```python
n_grid_points = 300

init_distribution = np.zeros(n_grid_points)


for i in range(30):
    
    init_distribution[i+1] = 1.

```

```python
generic_ca = GenericCoolingAcceleration(n_grid_points=n_grid_points,
                                        C0 = 1.,
                                        t_acc= 1E-4,
                                        cooling_index=2.,
                                        acceleration_index=2.,
                                        initial_distribution = init_distribution,
                                        store_progress = True
                                        
                                       
                                       )
```

```python
for i in range(50):
    
    generic_ca.solve_time_step()
```

```python
generic_ca.plot_evolution(skip=2);
```

```python
synch_cool = SynchrotronCooling(B=10.,
                                index=-2.2,
                                gamma_injection=1E3,
                                gamma_cool=200,
                                gamma_max=1E5,
                                store_progress=True
                               
                               )
```

```python
synch_cool.run()
```

```python
fig = synch_cool.plot_evolution(skip=50,
                                alpha=.4,
                                cmap='viridis',
                                reversed=True);
ax = fig.get_axes()[0]
ax.set_xlim(100)
ax.set_ylim(1E-10)
```
